#### Q4: Display a list of all text included within bullet points

```python
bullets = soup.find("div", {"class":"page_content"}).find_all("ul")
```

```python
for each in bullets:
    for list_ in (each.find_all("li")):
        print(list_.text)
```

#### Q5: Scrape a different page of your choice, parse the page and convert the information into a “tidy” pandas dataframe that can be used for further analysis.

```python
html = requests.get("https://www.apple.com").text
soup = BeautifulSoup(html, 'lxml')
```

```python
sections = soup.find("main", {"class":"main"}).find_all("section")
```

```python
sections2 = soup.find("main", {"class":"main", "data-module-template":"heroes"})#.find_all("section")


print(sections2)
```

```python
secs = []

for each in sections:
    for list_ in (each.find_all("div")):
#         print(list_.text)
        secs.append(list_.text)
#         print("BREAK!!!!")


print(secs)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(secs, columns=['Apple Homepage Headlines'])


df
```

```python
df['Apple Homepage Headlines'] = df['Apple Homepage Headlines'].str.strip()

df
```

```python
df = df.drop_duplicates()

df
```

```python
df = df.drop(9)


df
```

```python
df.reset_index(inplace=True, drop=True)
```
